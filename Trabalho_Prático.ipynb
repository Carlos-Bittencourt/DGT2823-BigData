```python
# TRABALHO PRÁTICO - Limpeza de Dados

import pandas as pd
import numpy as np

# Passo 3 e 4 - Ler CSV e atribuir à variável
df = pd.read_csv(
    'pico_web.csv',
    sep=';',
    engine='python',
    encoding='utf-8'
)

# Passo 5a - Informações gerais
print("Informações gerais:")
df.info(memory_usage='deep')

# Passo 5b - Primeiras e últimas 10 linhas
print("\nPrimeiras 10 linhas:")
print(df.head(10))

print("\nÚltimas 10 linhas:")
print(df.tail(10))


# Passo 6 - Cópia do dataset original
df_copy = df.copy()

# Passo 7a - Substituir NaN em Calories por 0
df_copy['Calories'] = df_copy['Calories'].fillna(0)

# Passo 7b - Verificar mudança
print("\nApós tratar Calories:")
print(df_copy)


# Passo 8a - Substituir NaN em Date por '1900/01/01'
df_copy['Date'] = df_copy['Date'].fillna('1900/01/01')

# Passo 8b - Verificar mudança
print("\nApós tratar Date:")
print(df_copy)


# Passo 9a - Substituir '1900/01/01' por NaN
df_copy['Date'] = df_copy['Date'].replace('1900/01/01', np.nan)

# Passo 9b - (Comentado pois ainda gera erro com valor inválido)
# df_copy['Date'] = pd.to_datetime(df_copy['Date'], format='%Y/%m/%d')


# Passo 10 - Corrigir valor específico '20201226'
df_copy['Date'] = df_copy['Date'].replace('20201226', '2020/12/26')


# Passo 11 - Converter Date para datetime (AGORA SEM ERRO)
df_copy['Date'] = df_copy['Date'].astype(str)
df_copy['Date'] = pd.to_datetime(df_copy['Date'], format='%Y/%m/%d', errors='coerce')

print("\nApós conversão para datetime:")
print(df_copy)


# Passo 12 - Remover linhas com valores nulos
df_copy = df_copy.dropna()


# Passo 13 - Resultado final
print("\nDataFrame final:")
print(df_copy)
```
